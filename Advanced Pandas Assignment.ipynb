**Q3 Customer Purchase Analysis**

**Identify the top 5 customers based on total sales.**

**Hint: Use GroupBy with aggregation**

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Advanced_pandas_sales_dataset.csv')

# Group by Customer and calculate total sales
customer_sales = df.groupby('Customer')['Sales'].sum().reset_index()

# Sort by Sales in descending order and get top 5
top_5_customers = customer_sales.sort_values(by='Sales', ascending=False).head(5)

print("Top 5 Customers based on Total Sales:")
print(top_5_customers)
```

**Q4 Regional Sales Performance**

**Calculate the total sales generated by each region.**

**Then identify:**

**Which region generated the highest revenue?**

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Advanced_pandas_sales_dataset.csv')

# Calculate total sales by region
regional_sales = df.groupby('Region')['Sales'].sum().reset_index()

# Find the region with the highest revenue
highest_revenue_region = regional_sales.loc[regional_sales['Sales'].idxmax()]

print("Total Sales by Region:")
print(regional_sales)
print("\nRegion with the highest revenue:")
print(highest_revenue_region)
```

**Q5 Product Performance Analysis**

**Calculate the total quantity sold for each product.**

**Which product appears to be the most popular?**

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Advanced_pandas_sales_dataset.csv')

# Calculate total quantity sold for each product
product_performance = df.groupby('Product')['Quantity'].sum().reset_index()

# Sort products by Quantity to find the most popular
product_performance_sorted = product_performance.sort_values(by='Quantity', ascending=False)

# Identify the most popular product
most_popular_product = product_performance_sorted.iloc[0]

print("Total Quantity Sold per Product:")
print(product_performance_sorted)

print("\nMost Popular Product:")
print(most_popular_product)
```

**Q6 Multi-Level Grouping**

**Group the dataset by: Region and Category**

**Calculate the total sales for each combination.**

**What insights can you derive from this analysis?**

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Advanced_pandas_sales_dataset.csv')

# Multi-level grouping: Region and Category
multi_group_sales = df.groupby(['Region', 'Category'])['Sales'].sum().reset_index()

# Sort for better readability
multi_group_sales = multi_group_sales.sort_values(by=['Region', 'Sales'], ascending=[True, False])

print("Total Sales by Region and Category:")
print(multi_group_sales)

# Pivot for better insight observation
pivot_sales = multi_group_sales.pivot(index='Region', columns='Category', values='Sales')
print("\nPivot Table View:")
print(pivot_sales)
```

**Q7 Pivot Table Creation**

**Create a pivot table showing:**

**Rows → Region**

**Columns → Product**

**Values → Total Sales**

**Which region sells the most laptops?**

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Advanced_pandas_sales_dataset.csv')

# Create the pivot table
# Rows -> Region, Columns -> Product, Values -> Sum of Sales
pivot_table = pd.pivot_table(df, values='Sales', index='Region', columns='Product', aggfunc='sum')

print("Pivot Table (Sales by Region and Product):")
print(pivot_table)

# Identify the region with the highest laptop sales
most_laptops_region = pivot_table['Laptop'].idxmax()
print(f"\nThe region that sells the most laptops is: {most_laptops_region}")
```

**Q8 Profitability Analysis**

**Calculate the average profit for each product category.**

**Which category is most profitable?**

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Advanced_pandas_sales_dataset.csv')

# Calculate the average profit for each category
category_profitability = df.groupby('Category')['Profit'].mean().reset_index()

# Sort the results by Profit in descending order
category_profitability = category_profitability.sort_values(by='Profit', ascending=False)

# Identify the most profitable category
most_profitable_category = category_profitability.iloc[0]

print("Average Profit per Category:")
print(category_profitability)

print(f"\nThe most profitable category (by average profit) is {most_profitable_category['Category']} "
      f"with an average profit of {most_profitable_category['Profit']:.2f}")
```

**Q9 Discount Impact Study**

**Analyze the relationship between discount and profit.**

**Tasks:**

**Calculate the average profit for each discount level**

**Identify whether higher discounts lead to lower profit**

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Advanced_pandas_sales_dataset.csv')

# Calculate the average profit for each discount level
discount_impact = df.groupby('Discount')['Profit'].mean().reset_index()

# Sort by Discount to observe the trend
discount_impact = discount_impact.sort_values(by='Discount')

print("Average Profit per Discount Level:")
print(discount_impact)

# Calculate correlation to quantify the relationship
correlation = df['Discount'].corr(df['Profit'])
print(f"\nCorrelation between Discount and Profit: {correlation:.4f}")

```

Conclusion: Yes, higher discounts significantly lead to lower profits in this dataset. While discounts might drive volume (sales quantity), they appear to be cutting deeply into the margins.

**Q10 End-to-End Data Wrangling Task**

**Perform the following workflow:**

**Filter the dataset where Sales > 3000**

**Group data by Region and Product**

**Calculate total sales**

**Reshape the result into a pivot table**

**Explain the business insights obtained from this analysis**

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Advanced_pandas_sales_dataset.csv')

# 1. Filter the dataset where Sales > 3000
filtered_df = df[df['Sales'] > 3000]

# 2 & 3. Group data by Region and Product and calculate total sales
grouped_sales = filtered_df.groupby(['Region', 'Product'])['Sales'].sum().reset_index()

# 4. Reshape the result into a pivot table
# Filling missing values (NaN) with 0 for better readability
pivot_table = grouped_sales.pivot(index='Region', columns='Product', values='Sales').fillna(0)

print("Final Pivot Table (Sales > 3000):")
print(pivot_table)
```
