**Q1. Create a countplot to show the number of passengers by gender.**

```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("Titanic.csv")

# Create the countplot
plt.figure(figsize=(6,4))
sns.countplot(x="Sex", data=df, palette="pastel")

# Add title and labels
plt.title("Number of Passengers by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")

plt.show()
```

**Q5. Create a scatterplot to**

 **a. show the relationship between age and fare.**




```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("Titanic.csv")

# Part (a) Scatterplot: Age vs Fare
plt.figure(figsize=(8,5))
sns.scatterplot(x="Age", y="Fare", data=df, alpha=0.6, color="blue")
plt.title("Relationship between Age and Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

 **b. analyze the relationship between family size (SibSp + Parch) and fare.**

```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("Titanic.csv")

# Part (b) Scatterplot: Family Size vs Fare
df["FamilySize"] = df["SibSp"] + df["Parch"]  # create family size column

plt.figure(figsize=(8,5))
sns.scatterplot(x="FamilySize", y="Fare", data=df, alpha=0.6, color="green")
plt.title("Relationship between Family Size and Fare")
plt.xlabel("Family Size (SibSp + Parch)")
plt.ylabel("Fare")
plt.show()
```

Interpretation
(a) Age vs Fare

The scatterplot shows no strong correlation between age and fare.

Fares are more influenced by passenger class than age.

Outliers are visible where some passengers (regardless of age) paid extremely high fares, typically 1st class tickets.

(b) Family Size vs Fare

Most passengers had small family sizes (0–3 relatives).

Larger families (5+) are rare and generally paid lower fares, likely because they traveled in 3rd class.

Higher fares are concentrated among small families or individuals, reflecting wealthier 1st class travelers.

**Q6. Create a countplot to show survival count by gender**

```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("Titanic.csv")

# Create the countplot
plt.figure(figsize=(7,5))
sns.countplot(x="Sex", hue="Survived", data=df, palette="Set1")

# Add title and labels
plt.title("Survival Count by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Survived", labels=["No", "Yes"])

plt.show()
```

**Q8. Create a heatmap to show correlation between numerical variables.**

```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("Titanic.csv")

# Select only numerical columns
numeric_df = df.select_dtypes(include=["float64", "int64"])

# Compute correlation matrix
corr_matrix = numeric_df.corr()

# Create the heatmap
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)

# Add title
plt.title("Correlation Heatmap of Numerical Variables")

plt.show()
```

**Q9. Identify which passenger class paid the highest fare on average.**

```python
# Import libraries
import pandas as pd

# Load the dataset
df = pd.read_csv("Titanic.csv")

# Calculate average fare by passenger class
avg_fare_by_class = df.groupby("Pclass")["Fare"].mean()

print("Average Fare by Passenger Class:")
print(avg_fare_by_class)
```

Interpretation
The output will show the average fare for 1st, 2nd, and 3rd class passengers.

From the Titanic dataset:

1st Class passengers paid the highest average fare.

2nd Class passengers paid a moderate fare.

3rd Class passengers paid the lowest average fare.
