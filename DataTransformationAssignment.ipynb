**Q1. Handle Missing Values (Basic Cleaning)**

**Fill missing values using the following:**

**Age → Median**

**Salary → Mean**

**Hours_Worked_Per_Week → Median**

**Performance_Score → Mean**

**Display the dataset after handling missing values.**

```python
import pandas as pd

# Load dataset
file = 'employee_productivity_dataset.csv'
df = pd.read_csv(file)

# Display first 5 rows
print(df.head())
print(df.info())


# Count missing values per column
print("count missing value",df.isnull().sum())

# Total missing values
print('Total missing:', df.isnull().sum().sum())

# Fill missing values
df['Age'].fillna(df['Age'].median())
df['Salary'].fillna(df['Salary'].mean())
df['Hours_Worked_Per_Week'].fillna(df['Hours_Worked_Per_Week'].median())
df['Performance_Score'].fillna(df['Performance_Score'].mean())

# Display cleaned dataset
print(df.head(12))
```

**Q7. Build Preprocessing Pipeline**

**Create a pipeline that:**

**Applies encoding to categorical columns**

**Applies scaling to numerical columns**

**Use:**

**ColumnTransformer**

**Pipeline**

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Define categorical and numerical columns
categorical_cols = ['Gender', 'Department', 'Work_Mode', 'Location']
numerical_cols = ['Age', 'Projects_Completed', 'Salary', 'Hours_Worked_Per_Week', 'Performance_Score']

# ColumnTransformer: OneHotEncoder for categorical, StandardScaler for numerical
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols),
        ('num', StandardScaler(), numerical_cols)
    ]
)

# Build pipeline
pipeline = Pipeline(steps=[('preprocessor', preprocessor)])

# Fit and transform dataset
df_processed = pipeline.fit_transform(df)

# Convert to DataFrame (no .toarray needed)
df_processed = pd.DataFrame(df_processed)

# Display first 12 rows
print(df_processed.head(12))
print("Transformed shape:", df_processed.shape)
```

**Q9. Conceptual Question**

**Why is scaling important in Python?**

**Ans:**  Scaling is crucial in Python because many machine learning algorithms are sensitive to the magnitude of features. Without scaling, models like SVMs, Logistic Regression, and KNN can perform poorly or converge slowly, while properly scaled data ensures fair feature contribution and faster, more accurate training.

Why Scaling Matters in Python
1. **Ensures Equal Feature Contribution**
Features like Salary (₹65,000) and Performance Score (4.5) exist on very different scales.

Without scaling, algorithms may give undue importance to large-scale features (like Salary) while ignoring smaller ones (like Performance Score).

Scaling ensures all features contribute equally.

2. **Improves Algorithm Performance**
Distance-based models (KNN, K-Means, PCA) rely on Euclidean distance. If one feature dominates due to scale, clustering or classification becomes biased.

Gradient-based models (Logistic Regression, Neural Networks) converge faster when features are standardized.

3. **Stabilizes Optimization**
Scaling reduces numerical instability during optimization.

For example, StandardScaler (Z-score normalization) makes features centered around mean = 0, std = 1, which helps gradient descent converge efficiently.

4. **Model-Specific Needs**
Tree-based models (Random Forest, XGBoost) are largely unaffected by scaling.

Linear models, SVMs, PCA, KNN require scaling for meaningful results.

Choosing the right scaler depends on the algorithm and dataset properties.

**Q10. Conceptual Question**

**Why do we convert categorical data into numerical form?**

**Ans:**  We convert categorical data into numerical form because most machine learning algorithms and statistical models work with numbers, not text. If we leave categories as strings (like Male/Female or HR/IT/Finance), the model cannot interpret them mathematically.

Reasons for Converting Categorical Data
1. **Model Compatibility**
Algorithms like Logistic Regression, SVM, KNN, Neural Networks require numerical inputs.

Text labels cannot be directly processed in mathematical equations.

2. **Mathematical Operations**
Models calculate distances, correlations, and weights.

For example, KNN uses Euclidean distance — it cannot compute distance between “HR” and “Finance” unless they are represented numerically.

3. **Avoiding Misinterpretation**
Assigning arbitrary numbers (HR = 1, IT = 2, Finance = 3) may imply an order that doesn’t exist.

Techniques like One-Hot Encoding or Label Encoding ensure categories are represented correctly without false hierarchy.

4. **Improves Accuracy**
Proper encoding allows the model to capture relationships between categorical features and target variables.

Example: Work_Mode (Remote/Hybrid/Onsite) encoded numerically helps predict productivity more effectively.

5. **Flexibility Across Algorithms**
Tree-based models (Random Forest, XGBoost) can handle categorical splits but still benefit from numerical encoding.

Linear models & distance-based models absolutely require numerical conversion.
