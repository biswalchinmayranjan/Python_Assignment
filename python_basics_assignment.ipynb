Question5. Write a Python program to check whether a number is even or odd using conditional statements.

```python
# Take input from the user
num = int(input("Enter a number: "))
# Check if the number is even or odd
if num % 2 == 0:
    print(f"{num} is Even")
else:
    print(f"{num} is Odd")
```

Question8.
Write a Python program to remove duplicate values from a list using a set.


```python
# Original list with duplicates
numbers = [2, 4, 6, 2, 8, 4, 10, 6]

# Convert list to set to remove duplicates
unique_numbers = list(set(numbers))

# Print result
print("Original list:", numbers)
print("List after removing duplicates:", unique_numbers)
```

Question 10 : Write a Python program to count how many times a number appears in a list.

Example List

[2,3,4,2,5,2]

```python
# Example list
numbers = [2,3,4,2,5,2]
# Dictionary to store frequencies
frequency = {}
# Loop through each number in the list
for num in numbers:
    if num in frequency:
        frequency[num] += 1
    else:
        frequency[num] = 1

# Print the result
print("Frequency of each number:")
for key, value in frequency.items():
    print(f"{key} appears {value} time(s)")
```
