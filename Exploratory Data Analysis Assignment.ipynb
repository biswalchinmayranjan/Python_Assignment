**Q1. What is the average, median, and standard deviation of house prices?**

```python
import pandas as pd

# Load CSV file
df = pd.read_csv("House Price India.csv")
df.columns = df.columns.str.replace(' ', '_')
# Quick checks
df.info()
df.describe()
```

```python
# Check for missing values
df.isnull().sum()

# Drop or impute if necessary
df = df.dropna(subset=['Price'])  # assuming 'Price' is the column name
```

```python
# Check for missing values
df.isnull().sum()

# Drop or impute if necessary
df = df.dropna(subset=['Price'])  # assuming 'Price' is the column name
```

```python
# Average (mean)
avg_price = df['Price'].mean()

# Median
median_price = df['Price'].median()

# Standard Deviation
std_price = df['Price'].std()

avg_price, median_price, std_price
print('Avg Price',avg_price)
print('Median Price',median_price)
print('Standard Deviation',std_price)
```

**Q2. Which number of bedrooms is most common?**

```python
# Frequency of each bedroom count
bedroom_counts = df['number_of_bedrooms'].value_counts()

# Most common bedroom count
most_common_bedrooms = bedroom_counts.idxmax()
most_common_count = bedroom_counts.max()

bedroom_counts, most_common_bedrooms, most_common_count
```

**Q3. Check if the price data is skewed.**

```python
# Skewness of house prices
price_skewness = df['Price'].skew()
price_skewness
```

Skewness > 0 (positive skew): Distribution is right-skewed → more houses with lower prices, but a few very expensive houses pull the tail to the right.

Skewness < 0 (negative skew): Distribution is left-skewed → more houses with higher prices, but a few very cheap houses pull the tail to the left.

Skewness ≈ 0: Distribution is approximately symmetric (close to normal).

**This means the house price distribution is positively skewed.**

**Q4. What is the average price for each number of bedrooms?**

```python
# Average price per bedroom category
avg_price_per_bedroom = df.groupby('number_of_bedrooms')['Price'].mean()

avg_price_per_bedroom
```

**Q5.  What is the relationship between living area and price? (Use : Area without Basement).**

```python
# Average price by living area (without basement)
area_price_relation = df.groupby('Area_of_the_house(excluding_basement)')['Price'].mean()

area_price_relation.head()  # show first few rows
```

```python
# Correlation between area and price
correlation = df['Area_of_the_house(excluding_basement)'].corr(df['Price'])
correlation
```

Interpretation
The groupby result shows how average price changes with different living area sizes.

The correlation value tells you the strength and direction of the relationship:

Positive correlation (>0): Larger living area → higher price.

Negative correlation (<0): Larger living area → lower price (unlikely in housing).

Near 0: Weak or no linear relationship.

**Correlation = 0.61 → strong positive relationship.**

**Q6. Identify any anomalies where houses have high prices but low areas**

```python
# Calculate thresholds
high_price_threshold = df['Price'].quantile(0.75)   # top 25% prices
low_area_threshold = df['Area_of_the_house(excluding_basement)'].quantile(0.25)  # bottom 25% areas

# Filter anomalies
anomalies = df[(df['Price'] > high_price_threshold) &
               (df['Area_of_the_house(excluding_basement)'] < low_area_threshold)]

anomalies[['Price', 'Area_of_the_house(excluding_basement)']].head()
```

**Q7. Compare average price based on number of floors and houses with or without waterfront.**

```python
# Average price per number of floors
avg_price_floors = df.groupby('number_of_floors')['Price'].mean()

avg_price_floors
```

```python
# Average price for waterfront vs non-waterfront houses
avg_price_waterfront = df.groupby('waterfront_present')['Price'].mean()

avg_price_waterfront
```

```python
# Average price grouped by floors and waterfront
avg_price_combined = df.groupby(['number_of_floors', 'waterfront_present'])['Price'].mean()

avg_price_combined
```

**Q8. Identify the minimum and maximum house price. What does this indicate?**

```python
# Minimum and maximum house price
min_price = df['Price'].min()
max_price = df['Price'].max()

print('Min Price',min_price)
print('Max Price',max_price)
```

Business Insight
The wide gap between minimum and maximum prices indicates a highly diverse housing market.

This suggests that location, amenities, and property type play a huge role in pricing.

For analysis, it’s important to note that extreme values (outliers) can skew averages. That’s why median price (from Q1) is often more reliable for understanding the “typical” market.

Developers and agents can segment the market into affordable, mid-range, and luxury categories to better target buyers.

**Q9. Which location (zipcode/area) has the highest average price?**

```python
# Average price per zipcode
avg_price_zipcode = df.groupby('Postal_Code')['Price'].mean()

# Find the zipcode with the highest average price
highest_avg_price_zipcode = avg_price_zipcode.idxmax()
highest_avg_price_value = avg_price_zipcode.max()

print('Zipcode',highest_avg_price_zipcode)
print('Price',highest_avg_price_value)
```

Business Insight
The location with the highest average price is likely a premium area (e.g., city center, waterfront, or upscale neighborhood).

This indicates that location is a major driver of house prices, sometimes even more than size or number of bedrooms.

Developers and agents should focus marketing efforts on these high-value areas, while buyers may need to balance affordability with location desirability.
