Q1. Creating a DataFrame

Create a Pandas DataFrame using a Python dictionary with the following columns:

Product

Sales

Q4. Indexing using .loc
Using .loc, perform the following:

Display the row with index 10.

Display the Sales value of the row with index 20

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("Retail_dataset.csv")

# 1. Display the row with index 10
row_10 = df.loc[10]
print("Row with index 10:")
print(row_10)

# 2. Display the Sales value of the row with index 20
sales_20 = df.loc[20, "Sales"]
print("\nSales value at index 20:")
print(sales_20)
```

Q5. Indexing using .iloc

Using .iloc, display:

The first 6 rows and first 4 columns.

The Sales value from the 3rd row.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("Retail_dataset.csv")

# 1. Display the first 6 rows and first 4 columns
subset = df.iloc[0:6, 0:4]
print("First 6 rows and first 4 columns:")
print(subset)

# 2. Display the Sales value from the 3rd row
sales_3rd = df.iloc[2, 4]   # row index 2 (3rd row), column index 4 (Sales)
print("\nSales value from the 3rd row:")
print(sales_3rd)
```

Q7. Filtering Data

Filter the dataset to display:

Rows where Region = "East".

Rows where Sales > 1500.


```python
import pandas as pd

# Load the dataset
df = pd.read_csv("Retail_dataset.csv")

# 1. Rows where Region = "East"
east_region = df[df["Region"] == "East"]
print("Rows where Region = 'East':")
print(east_region.head(10))   # showing first 10 rows for preview

# 2. Rows where Sales > 1500
high_sales = df[df["Sales"] > 1500]
print("\nRows where Sales > 1500:")
print(high_sales.head(10))    # showing first 10 rows for preview
```

Q8. Multiple Condition Filtering

Display rows where:

Region is West AND Sales is greater than 1200

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("Retail_dataset.csv")

# Apply multiple condition filtering
filtered_data = df[(df["Region"] == "West") & (df["Sales"] > 1200)]

# Display the result
print("Rows where Region = 'West' AND Sales > 1200:")
print(filtered_data.head(10))   # showing first 10 rows for preview
```

Q9. Descriptive Statistics

Generate summary statistics using:

describe()

From the result identify:

Mean sales value

Maximum quantity sold

Minimum sales value

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("Retail_dataset.csv")

# Generate summary statistics
stats = df.describe()
print("Summary Statistics:")
print(stats)

# Extract specific values
mean_sales = stats.loc["mean", "Sales"]
max_quantity = stats.loc["max", "Quantity"]
min_sales = stats.loc["min", "Sales"]

print("\nMean Sales Value:", mean_sales)
print("Maximum Quantity Sold:", max_quantity)
print("Minimum Sales Value:", min_sales)
```

Q10. Aggregation Analysis

Using the Sales column, calculate:

Total sales

Average sales

Maximum sales

Minimum sales

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("Retail_dataset.csv")

# Aggregation calculations
total_sales = df["Sales"].sum()
average_sales = df["Sales"].mean()
max_sales = df["Sales"].max()
min_sales = df["Sales"].min()

print("Total Sales:", total_sales)
print("Average Sales:", average_sales)
print("Maximum Sales:", max_sales)
print("Minimum Sales:", min_sales)
```
